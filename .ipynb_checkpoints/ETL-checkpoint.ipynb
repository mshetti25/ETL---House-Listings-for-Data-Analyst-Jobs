```python
import pandas as pd
from sqlalchemy import create_engine
```

## Store CSV into DataFrames

```python
csv_file = "Resources/Housing_Data.csv"
housing_df = pd.read_csv(csv_file)
housing_df.describe()
```

```python
csv_file = "Resources/Data_Analyst.csv"
DA_df = pd.read_csv(csv_file)
DA_df.describe()
```

```python
housing_df.head()
```

## Cleaning Data

### Shortlisting final columns

```python
housing_df = housing_df[["id", "City", "price", "sqfeet", "state" ]]
housing_df
```

```python
housing_df = housing_df.dropna()
housing_df
```

```python
DA_df["City"].unique()
```

```python
housing_df["City"].unique()
```

### Cleaning primary key values

```python
housing_df["City"] = housing_df["City"].replace("SF bay area", "San Francisco")
```

```python
housing_df["City"].unique()
```

```python
housing_df["City"] = housing_df["City"].replace("new york city", "New York")
housing_df["City"] = housing_df["City"].replace("chicago", "Chicago")
housing_df["City"] = housing_df["City"].replace("san diego", "San Diego")
housing_df["City"] = housing_df["City"].replace("seattle-tacoma", "Seattle")
```

```python
housing_df["City"].unique()
```

### Finding Average Rental Price per Sq. Feet

```python
housing_df["Rent Price/Sq.Ft"] = round(housing_df["price"]/housing_df["sqfeet"],2)
```

```python
house_df = housing_df[["City", "price", "sqfeet", "state", "Rent Price/Sq.Ft"]]
```

```python
house_df
```

```python
final_house_df = house_df.groupby("City").mean("Rent Price/Sq.Ft")
```

```python
final_house_df["id"] = [1, 2, 3, 4, 5]
```

```python
# final_house_df = final_house_df[["City", "id", "Rent Price/Sq.Ft"]]
final_house_df
```

```python
final_house_df = final_house_df.drop(columns=['sqfeet'])
```

```python
final_house_df
```

```python
final_house = final_house_df.round({'price' : 2, "Rent Price/Sq.Ft" : 2})
```

```python
final_house.reset_index()
```

```python
final_house = final_house.reset_index()
```

```python
final_house
```

```python
final_housing = final_house.set_index("id")
```

```python
final_housing
```

```python
data_analyst_df = DA_df.drop(columns=['Rating', 'Location', 'Founded', 'Sector', 'Competitors'])
data_analyst_df
```

### Connect to local database

```python
rds_connection_string = "username:password@localhost:5432/DAHouse_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
final_housing.to_sql(name='housings', con=engine, if_exists='append', index=False)
```

### Confirm data has been added by querying the housing table

```python
pd.read_sql_query('select * from housings', con=engine).head()
```

```python
data_analyst_df.to_sql(name='jobs', con=engine, if_exists='append', index=False)
```

### Confirm data has been added by querying the jobs table

```python
engine.table_names()
```

```python
pd.read_sql_query('select "Salary Estimate" from jobs', con=engine).head()
```

### Jobs in Chicago sorted by Salary (High to Low)

```python
pd.read_sql_query('select jobs.*, housings.price, housings."Rent Price/Sq.Ft" from jobs join housings on jobs."City" = housings."City" where jobs."City"=\'Chicago\' order by jobs."Salary Estimate", jobs."Job Title"', con=engine).head()
```

### NonProfit Organization Jobs

```python
pd.read_sql_query('select jobs.*, housings.price, housings."Rent Price/Sq.Ft" from jobs join housings on jobs."City" = housings."City" where jobs."Type of ownership" = \'Nonprofit Organization\' order by jobs."City",jobs."Salary Estimate", jobs."Job Title"', con=engine)
```

### Top Salaried Jobs

```python
pd.read_sql_query('select jobs.*, housings.price, housings."Rent Price/Sq.Ft" from jobs join housings on jobs."City" = housings."City" order by jobs."Salary Estimate" desc, jobs."Job Title"', con=engine).head(20)
```
